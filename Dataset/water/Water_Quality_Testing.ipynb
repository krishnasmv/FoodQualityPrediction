```python
print('splitting the dat into training and testing')
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state = 42,stratify=y)
print("Training size = ", X_train.shape)
print("Testing size = ", X_test.shape)
```

Defining models

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    "Random Forest": RandomForestClassifier(n_estimators=50, max_depth=5),
    "KNN": KNeighborsClassifier(n_neighbors=8),
    "SVM": SVC(kernel='rbf', C=1.0),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss", max_depth=4, n_estimators=50)
}
```

Train and evaluate

```python

accuracies = {}
print("\nModel Accuracies:")
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    accuracies[name] = acc
    print(f"{name}: {acc:.2%}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sorted_items = sorted(accuracies.items(), key=lambda x: x[1], reverse=True)
model_names = [name for name, acc in sorted_items]
model_scores = [acc for name, acc in sorted_items]

y_min = max(0, min(model_scores) - 0.05)
y_max = min(1.0, max(model_scores) + 0.05)
```

```python
plt.figure(figsize=(10, 6))
bar_colors = sns.color_palette("crest", len(model_names))
bars = sns.barplot(x=model_names, y=model_scores, palette=bar_colors)

plt.title("Accuracy of ML Models", fontsize=16, fontweight='bold')
plt.ylabel("Accuracy", fontsize=12)
plt.xticks(rotation=25, fontsize=11)
plt.ylim(y_min, y_max)
plt.grid(axis='y', linestyle='--', alpha=0.5)

for bar, acc in zip(bars.patches, model_scores):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 0.005,
             f"{acc:.2%}",
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```
